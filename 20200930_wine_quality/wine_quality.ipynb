## クラスター分析（クロス集計）

```python
from sklearn.decomposition import PCA
from sklearn import cluster
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
wine_red_cls = wine_red[['遊離二酸化硫黄', '総二酸化硫黄']].copy()
clustering_sc = sc.fit_transform(wine_red_cls)

kmeans = KMeans(n_clusters=2, random_state=0)
clusters = kmeans.fit(clustering_sc)
wine_red_cls['cluster'] = clusters.labels_

pca = PCA(n_components=2)
x = clustering_sc
pca.fit(x)
x_pca = pca.transform(x)
pca_df = pd.DataFrame(x_pca)
pca_df['cluster'] = wine_red_cls['cluster']

for i in wine_red_cls['cluster'].unique():
    tmp = pca_df.loc[pca_df['cluster'] == i]
    plt.scatter(tmp[0], tmp[1])

plt.grid()
plt.show();
```

```python
# 非階層クラスタリング
## クロス集計
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn import cluster
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()

wine_red_cls = wine_red[['固定酸度', '揮発性酸性度']].copy()
clustering_sc = sc.fit_transform(wine_red_cls)

kmeans = KMeans(n_clusters=2, random_state=0)
clusters = kmeans.fit(clustering_sc)
wine_red_cls['cluster'] = clusters.labels_

x = clustering_sc
pca = PCA(n_components=2)
pca.fit(x)
x_pca = pca.transform(x)
pca_df = pd.DataFrame(x_pca)
pca_df['cluster'] = wine_red_cls['cluster']

for i in wine_red_cls['cluster'].unique():
    tmp = pca_df.loc[pca_df['cluster'] == i]
    plt.scatter(tmp[0], tmp[1])
plt.grid()
plt.show();
```

## スミルノフ=グラブス検定（外れ値検定）

```python
# スミルノフ=グラブス検定
data = wine_red['総二酸化硫黄']
alpha = 0.5

x, o = list(data), []
while True:
    n = len(x)
    t = stats.t.isf(q=(alpha / n) / 2, df=n - 2)
    tau = (n - 1) * t / np.sqrt(n * (n - 2) + n * t * t)
    i_min, i_max = np.argmin(x), np.argmax(x)
    myu, std = np.mean(x), np.std(x, ddof=1)
    i_far = i_max if np.abs(x[i_max] - myu) > np.abs(x[i_min] - myu) else i_min
    tau_far = np.abs((x[i_far] - myu) / std)
    if tau_far < tau: break
    o.append(x.pop(i_far))

not_out = np.array(x)
out = np.array(o)
out.astype('int').tolist()
```

## ホテリング理論

```python
# カイ二乗分布
x = np.linspace(0, 8, 1000)
fig,ax = plt.subplots(1,1, figsize=(10,5))

linestyles = [':', '--', '-.', '-']
deg_of_freedom = [1] # , 2, 3, 4
for k, ls in zip(deg_of_freedom, linestyles):
  ax.plot(x, stats.chi2.pdf(x, k), linestyle=ls, label=r'$k=%i$' % k)

plt.xlim(0, 8)
plt.ylim(0, 1.0)
plt.grid()
plt.legend()
plt.show();
```

```python
# カイ二乗分布
x = np.linspace(0, 8, 1000)
fig,ax = plt.subplots(1,1, figsize=(10,5))
 
linestyles = [':', '--', '-.', '-']
deg_of_freedom = [1, 2, 3, 4]
for k, ls in zip(deg_of_freedom, linestyles):
  ax.plot(x, stats.chi2.pdf(x, k), linestyle=ls, label=r'$k=%i$' % k)
 
plt.xlim(0, 8)
plt.ylim(0, 1.0)
plt.grid()
plt.legend()
plt.show();
```

```python
x = np.linspace(0.000001, 8, 1000)
fig,ax = plt.subplots(1,1, figsize=(7,7))

df = (1,1)
ls = '-'
y = stats.f.pdf(x, df[0], df[1])
ax.plot(x, y, linestyle=ls, label=f'k = {df[0]}')

plt.xlim(0, 8)
plt.ylim(0, 1.0)
plt.fill_between(x, y, 0, where=x>=3.84, facecolor='r',alpha=0.5)
plt.title('自由度１のカイ二乗分布')
plt.axvline(3.84, c='blue', ls='--')
plt.axvline(4, c='red', ls='-.')
plt.grid()
plt.legend()
plt.show();
```

## 標準正規分布の描画（考察用）
